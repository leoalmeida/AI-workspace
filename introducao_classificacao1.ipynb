```python
uri_tracking = 'https://gist.github.com/guilhermesilveira/b9dd8e4b62b9e22ebcb9c8e89c271de4'
uri_projetos = 'https://gist.github.com/guilhermesilveira/12291c548acaf544596795709020e3db'

#features [1 sim, 0 não]

porco1 = [0,1,0]
porco2 = [0,1,1]
porco3 = [1,1,0]
cachorro1 = [0,1,1]
cachorro2 = [1,0,1]
cachorro3 = [1,1,1]

# 1 -> porco, 0 -> cachorro
treino_x = [porco1, porco2, porco3, cachorro1, cachorro2, cachorro3]
treino_y = [1,1,1,0,0,0]
```

```python
from sklearn.svm import LinearSVC

modelo = LinearSVC()
modelo.fit(treino_x, treino_y)
```

```python
animal_misterio = [0,0,0]
modelo.predict([animal_misterio])
```

```python
misterio1 = [1,1,1]
misterio2 = [1,1,0]
misterio3 = [0,1,1]
teste_x = [misterio1,misterio2, misterio3]
previsoes = modelo.predict(teste_x)
```

```python
teste_y = [0,1,1]
```

```python
corretos = (previsoes == teste_y).sum()
total = len(teste_x)
taxa_acerto = (corretos/total) * 100

print(f"Acurácia {taxa_acerto:.2f}%")
```

```python
from sklearn.metrics import accuracy_score

taxa_de_acerto = accuracy_score(teste_y, previsoes) * 100
print(f"Acurácia {taxa_de_acerto:.2f}%")
```
